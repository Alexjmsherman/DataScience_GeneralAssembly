```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#get automobile mpg data
data = pd.read_table(r'https://raw.githubusercontent.com/justmarkham/DAT5/master/data/auto_mpg.txt', sep='|') 

data.head()
```

### Part 1

```python
#Produce a plot that compares the mean mpg for the different # of cylinders.
data.groupby('cylinders')['mpg'].mean().plot(kind='bar', \
    title='mpg is greatest at 4 cylinders and is skewed to the right')
plt.xlabel('Cylinders')
plt.ylabel('Mean mpg')

#The plot show that 
```

### Part 2

```python
#Use a scatter matrix to explore relationships between different numeric variables.

data.plot(kind='scatter', x='weight', y='horsepower', 
          c='model_year', colormap='Blues', alpha=0.25)

# The plot displays that over the period from the 1970's to early 1980's.
# the cars in the dataset trended towards less horsepower and weight 
# in a linear fashion with descreasing variance over time
```

### Part 3

```python
#Do heavier or lighter cars get better mpg?

data.groupby(data.weight > data.weight.median())['mpg'].mean() \
    .plot(kind='bar', title='Lighter cars get better mpg')
plt.xlabel('Weight greater than the Median')
plt.ylabel('MPG')
```

```python
#How are horsepower and displacement related?

data.plot(kind='scatter', x='displacement', y='horsepower', alpha=0.5, \
          title='Horsepower & displacement have a positive linear correlation')
```

```python
#-What does the distribution of acceleration look like?

data.acceleration.hist(bins=25)
plt.title("Accerlation is slightly skewed to the right")
plt.xlabel('Accelation')
plt.ylabel('Frequency')
```

```python
#How is mpg spread for cars with different numbers of cylinders?

data['mpg'].hist(by=data.cylinders, sharex=True, sharey=True, \
                 bins=10, layout=(5,1))

#for common cylinder amounts, mpg spread decreases as cylinders increase
```

```python
#Do cars made before or after 1975 get better average mpg?

data.groupby(data.model_year < 75)['mpg'].mean() \
    .plot(kind='bar', title='Cars created after 1975 get better average mpg')
plt.xlabel('Was the car created before 1975')
plt.ylabel('Mean mpg')
```
